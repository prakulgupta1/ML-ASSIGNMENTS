# Part III: Calculating Proximity /Correlation Analysis of two features
Make sure each attribute is transformed in a same scale for numeric attributes and Binarization for each nominal attribute, and each discretized numeric attribute to standardization. Make sure to apply a correct similarity measure for nominal (one hot encoding)/binary attributes and numeric attributes respectively.

(a) Calculate Similarity in Simple Matching, Jaccard Similarity, and Cosine Similarity between two following objects of your transformed input data.

(b) Calculate Correlation between two features Commute Distance and Yearly Income

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import jaccard
import numpy as np

# Load the data

df = data.copy()  # Create a copy of the original DataFrame

# Select numeric and categorical columns
numeric_cols = ['YearlyIncome', 'NumberCarsOwned', 'NumberChildrenAtHome', 'TotalChildren', 'AvgMonthSpend']
categorical_cols = ['Education', 'Occupation', 'MaritalStatus', 'Gender', 'HomeOwnerFlag']

# Step 1: Standardize numeric attributes
scaler = StandardScaler()
numeric_data = scaler.fit_transform(df[numeric_cols])

# Step 2: Binarize categorical attributes (One-Hot Encoding)
encoder = OneHotEncoder(sparse_output=False)
categorical_data = encoder.fit_transform(df[categorical_cols])

# Combine the processed data
processed_data = np.hstack([numeric_data, categorical_data])

# Convert processed data back to a DataFrame for easier handling
processed_df = pd.DataFrame(processed_data)

# Step 3: Calculate Similarity between two objects (first two rows as an example)
obj1 = processed_df.iloc[0].values
obj2 = processed_df.iloc[1].values

# Simple Matching Coefficient (SMC)
def simple_matching_coefficient(a, b):
    return sum(a == b) / len(a)

smc = simple_matching_coefficient(obj1, obj2)
print("Simple Matching Coefficient:", smc)

# Jaccard Similarity
def jaccard_similarity(a, b):
    return 1 - jaccard(a, b)

jacc_sim = jaccard_similarity(obj1, obj2)
print("Jaccard Similarity:", jacc_sim)

# Cosine Similarity
cos_sim = cosine_similarity([obj1], [obj2])[0][0]
print("Cosine Similarity:", cos_sim)

# Step 4: Calculate Correlation between `AvgMonthSpend` and `YearlyIncome`
correlation = df['AvgMonthSpend'].corr(df['YearlyIncome'])
print("Correlation between AvgMonthSpend and YearlyIncome:", correlation)
```
